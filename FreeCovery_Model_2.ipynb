```python
df= pd.read_csv('Addictions.csv')
```

```python

data=df[["cleaned","scores"]]
data
```

```python
import json 
```

```python
train_y = data["scores"][:].tolist()

# Replace single quotes with double quotes
train_y = [x.replace("'", '"') for x in train_y]

# Convert stringified dictionaries to actual dictionaries
train_y = [json.loads(x) for x in train_y]

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['sentiment'])
```

```python
for index in train_y:
    if index['pos'] > index['neg']:
        result_df = result_df.append({'sentiment': 1}, ignore_index=True)
    else:
        result_df = result_df.append({'sentiment': 0}, ignore_index=True)

# Print the result DataFrame
print(result_df)
```

```python
concatenated_df = pd.concat([data["cleaned"][:], result_df], axis=1)
concatenated_df
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForPreTraining
from torch.utils.data import DataLoader, Dataset

# Define the pre-training dataset
class PretrainingDataset(Dataset):
    def __init__(self, dataframe, tokenizer):
        self.data = dataframe
        self.tokenizer = tokenizer
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        text = self.data.iloc[idx]['cleaned']
        label = self.data.iloc[idx]['sentiment']
        
        inputs = self.tokenizer(text, truncation=True, padding=True)
        inputs['sentiment'] = label
        
        return inputs

# Load the pre-training data and initialize the model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
#dataframe = pd.read_csv('dataframe.csv') # replace with your own data
dataset = PretrainingDataset(concatenated_df, tokenizer)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

model = BertForPreTraining.from_pretrained('bert-base-uncased')

# Pre-train the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)

for epoch in range(3):
    for batch in dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        labels = batch['sentiment'].to(device)
        
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            labels=labels
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
    
    # Save the model after each epoch
    model.save_pretrained(f'pretrained_bert_epoch{epoch+1}.pt')
```
