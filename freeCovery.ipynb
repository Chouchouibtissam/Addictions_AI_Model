```python
df= pd.read_csv('Addictions.csv')
```

```python

data=df[["cleaned","scores"]]
data
```

```python
import json 
```

```python
#type(f"""{data["scores"]}""")
```

```python
train_y = data["scores"][:].tolist()

# Replace single quotes with double quotes
train_y = [x.replace("'", '"') for x in train_y]

# Convert stringified dictionaries to actual dictionaries
train_y = [json.loads(x) for x in train_y]

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['sentiment'])
```

```python
for index in train_y:
    if index['pos'] > index['neg']:
        result_df = result_df.append({'sentiment': 1}, ignore_index=True)
    else:
        result_df = result_df.append({'sentiment': 0}, ignore_index=True)

# Print the result DataFrame
print(result_df)
```

```python
concatenated_df = pd.concat([data["cleaned"][:], result_df], axis=1)
concatenated_df
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Preprocess the cleaned data
concatenated_df['cleaned'] = concatenated_df['cleaned'].str.lower().str.replace('[^\w\s]', '')
```

```python
concatenated_df
```

```python
# Convert the text data into numerical features
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(concatenated_df['cleaned'].values.astype('U'))
y = concatenated_df['sentiment'].astype('int')
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Train the sentiment analysis model using Naive Bayes
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
# Test the model's accuracy on the testing data
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
